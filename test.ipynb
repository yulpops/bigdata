```python
import pandas as pd
from sklearn.model_selection import cross_val_score, KFold
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# загрузить данные из файла "dat.csv"
data = pd.read_csv("dat.csv")
data=data.drop('Total Deaths',axis=1)
# разделить данные на матрицу объектов признаков X и целевой вектор y
X = data.drop("Illness ",axis=1)
y = data["Illness "]

# создать объект модели линейной регрессии
model = LinearRegression()

# выполнить кросс-валидацию
cv = KFold(n_splits=10, shuffle=True, random_state=0)
scores = cross_val_score(model, X, y, cv=cv, scoring='r2')

# вычислить среднее значение метрики R2-score
mean_score = scores.mean()

print("Mean R2-score: {:.20f}".format(mean_score))
#list(data)
```

```python
import pandas as pd
from sklearn.model_selection import KFold, cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder

# Загрузка данных
data = pd.read_csv('dat.csv')
data=data.drop('Total Deaths',axis=1)
# Разделение признаков и целевой переменной
X = data.drop('Illness ', axis=1)
y = data['Illness ']

# One-hot encoding для категориальных признаков
categorical_features = X.select_dtypes(include=['object']).columns.tolist()
ohe = OneHotEncoder(sparse=False, handle_unknown='ignore')
X_categorical = pd.DataFrame(ohe.fit_transform(X[categorical_features]))
X_categorical.columns = ohe.get_feature_names(categorical_features)

# Объединяем one-hot encoded признаки и числовые признаки
X = pd.concat([X.drop(categorical_features, axis=1), X_categorical], axis=1)

# Создаем объект модели
model = LinearRegression()

# Определяем кросс-валидацию
cv = KFold(n_splits=10, shuffle=True, random_state=0)

# Вычисляем качество модели
mean_score = cross_val_score(model, X, y, cv=cv, scoring='r2').mean()

print("Mean R2-score:", mean_score)
```
